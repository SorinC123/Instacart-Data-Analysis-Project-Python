## Step 5. Combine the 'orders_products_combined' dataframe with the products data set.

```python
# Load the clean product file 'products_checked_v2.csv'
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked_v2.csv'), index_col = False)
```

```python
# Verify the contents of df_prods
df_prods.head()
```

```python
# I want to get rid of the original '_merge' column from 'df_orders_products_combined' dataframe
df_orders_products_combined = df_orders_products_combined.drop (columns = ['_merge'])
df_orders_products_combined
```

```python
# Combine the orders and the products dataframes on the 'product_id' column.
df_ords_prods_merge = df_orders_products_combined.merge(df_prods, on = ['product_id'], indicator = True)
```

```python
# Verify the contents of the merged dataframe
df_ords_prods_merge
```

```python
df_ords_prods_merge.shape
```

## Step 6. Confirm the results of the merge using the merge flag.

```python
# We notice that the number of rows in the merged dataframe is smaller than the number
# in the 'df_orders_products_combined' dataframe.
# To determine how many rows were left out, we need first to do a merge with an outer join.
df_ords_prods_merge_outer = df_orders_products_combined.merge(df_prods, on = ['product_id'], how = 'outer', indicator = True)
df_ords_prods_merge_outer.head()
```

```python
# Check the values of the '_merge' column
df_ords_prods_merge_outer['_merge'].value_counts()
```

```python
# From the results above, we determined that the merged dataframe has 32,399,732 rows which matched both dataframes
# ('df_orders_products_combined' and 'df_prods'), now in 'df_ords_prods_merge' dataframe.
# 35,327 rows were excluded from 'df_orders_products_combined' dataframe, and
# 11 rows were excluded from 'df_prods' dataframe.
```
