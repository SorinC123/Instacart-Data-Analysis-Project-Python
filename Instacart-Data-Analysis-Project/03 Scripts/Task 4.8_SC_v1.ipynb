## Step 4. Create a loyalty flag for existing customers using the transform() and loc() functions.

```python
# 1. Split the data into groups based on the “user_id” column.
# 2. Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
# 3. Create a new column, “max_order,” into which place the results of the aggregation.

df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head(15)
```

```python
# Create and populate the loyalty_flag
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check if the right flags were assigned
df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
# Print the frequency of the new “loyalty_flag”
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

## Step 5. Check the basic statistics of the product prices for each loyalty category.

```python
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['median', 'mean', 'min', 'max', 'std']})
```

```python
# We notice that the median values of the prices for the three groups is identical and the mean values are slightly different.
# This means that the prices of products purchased by loyal customers DO NOT differ from those purchased by regular or new customers.
```

## Step 6.  Create a spending flag for each user based on the average price across all their orders.

```python
# Create a new column for the average price of all orders of a customer
df_ords_prods_merge['average_price_all_orders'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge.head(60)
```

```python
# Assign the spender flag to each customer
df_ords_prods_merge.loc[df_ords_prods_merge['average_price_all_orders'] < 10,'spending_flag']='Low_spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_price_all_orders'] >= 10,'spending_flag']='High_spender'
```

```python
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge.head(60)
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# It was probably expected that the number of high spenders would be less than the number of low spenders.
```

## Step 7. Determine frequent versus non-frequent customers.

```python
# Create a new column for the median value in the “days_since_prior_order” column.
df_ords_prods_merge['order_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge.head()
```

```python
# Assign the order frequency flag to each customer according to the requirements

df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] > 20,'frequency_flag']='Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['order_frequency'] > 10) & (df_ords_prods_merge['order_frequency'] <= 20),'frequency_flag']='Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] <= 10,'frequency_flag']='Frequent customer'
```

```python
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merge[['user_id', 'order_frequency', 'frequency_flag']].head(60)
```

```python
# There are many more Frequent customers than Regular and Non-frequent customers.
```
