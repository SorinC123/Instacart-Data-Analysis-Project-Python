### Step 1.a. Create 'price_label' column using a user defined function

```python
# Create 'price_label' function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function 'price_label' on every row in the 'df_ords_prods_merge_1mil' dataframe.
df_ords_prods_merge_1mil['price_range'] = df_ords_prods_merge_1mil.apply(price_label, axis=1)
```

### Step 1.c. Create and populate the 'busiest day' column 

```python
# Calculate the frequency for the 'orders_day_of_week' ; will consider 0 as Monday, 6 as Sunday, and the others in between 0 and 6.
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create a list with the 3 values below, based on the day of the week
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# Assign the values in the result list to a new column, named 'busiest day', in the 'df_ords_prods_merge' dataframe.
df_ords_prods_merge['busiest day']= result
```

```python
# Check the frequency of the values for the 'busiest day' column.
df_ords_prods_merge['busiest day'].value_counts(dropna = False)
```

## Step 2. Create a new column to house the values 'Busiest Days' and 'Least busy', which will have a different meaning.

```python
# Create a new list and populate it according to the new requirements
result_1 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_1.append("Busiest days")
  elif value == 4 or value == 3:
    result_1.append("Slowest days")
  else:
    result_1.append("Regularly busy")
```

```python
# Verify the contents of the new list
result_1
```

```python
# Assign the values in the result list to a new column, named 'busiest days', in the 'df_ords_prods_merge' dataframe.
df_ords_prods_merge['busiest days']= result_1
```
