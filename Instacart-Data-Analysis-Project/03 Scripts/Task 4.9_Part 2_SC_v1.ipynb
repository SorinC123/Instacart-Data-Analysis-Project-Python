## Step 7. Is there a connection between age and spending power (income)?

```python
# Create data frame df_5 which contains only the columns we are interested in
df_5 = df_2[['age','income']]
```

```python
df_5.head()
```

```python
# Create a scatter plot.
scatterplot_age_income = sns.scatterplot(x='age', y='income', data=df_5)
```

```python
# We notice that for ages 20 to 40, the income is distributed up to $40,000 (and not more than this). 
# Starting with age 40, income levels are increasing, with more data points showing in higher income brackets. 
# The density of the data points for ages 40 and younger shows that the majority in this group have an income between $0 and $20,000,
# with some individuals earing up to $40,000.
# In the age group 40 and older, the majority is making up to $30,000 with a moderate number of people having an income between
# $30,000 and $60,000.
# This pattern is suggesting that the income increases with age. 
```
